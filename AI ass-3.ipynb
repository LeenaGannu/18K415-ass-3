```
#Read mobile dataset using pandas, and preprocess it (fill null values).
 
import pandas as pd 
ms=pd.read_excel('/content/drive/My Drive/mobile_cleaned-1549119762886_-452616671.csv')
print(ms)
ms.describe()
```

```
ms.isnull()
```

```
ms.isnull().sum()
```

```
ms['weight'] = ms['weight'].fillna(ms['weight'].mean())
ms['primary_camera_resolution'] = ms['primary_camera_resolution'].fillna(ms['primary_camera_resolution'].mean())
ms['video_resolution'] = ms['video_resolution'].fillna(ms['video_resolution'].mean()) 
ms['talk_time'] = ms['talk_time'].fillna(ms['talk_time'].mean())
ms['phone_height'] = ms['phone_height'].fillna(ms['phone_height'].mean())
ms['no_of_reviews_in_gsmarena_in_week'] = ms['no_of_reviews_in_gsmarena_in_week'].fillna(ms['no_of_reviews_in_gsmarena_in_week'].mean())
ms['stand_by_time'] = ms['stand_by_time'].fillna(ms['stand_by_time'].mean())
```

```
#Compute outer product of 2 vectors.
#Eg:
 
#original matrix:
#[[1, 0], [0, 1]]
#[[1, 2], [3, 4]]
#Outer product of the said two vectors:
#[[1 2 3 4]
# [0 0 0 0]
# [0 0 0 0]
# [1 2 3 4]]
 
import numpy as np
m1= np. array ([[1,0],[0,1]])
m2= np. array ([[1,2],[3,4]])
print ("original matrix ")
print(m1)
print(m2)
print("outer product of two matrices ")
result=np.outer(m1,m2)
print(result)
```

```
#Write a NumPy program to compute the condition number of a given matrix.
#Original matrix:
#[[1 2]
# [3 4]]
#Condition number of the said matrix:
#14.9330343737
 
import numpy as np
m = np.array([[1,2],[3,4]])
print("Original matrix:")
print(m)
result =  np.linalg.cond(m)
print("Condition number of the said matrix:")
print(result)
```

```
#write a NumPy program to compute the sum of the diagonal element of a given array(matrix).
 
import numpy as np
m = np.arange(9).reshape(3,3)
print("Original matrix:")
print(m)
result =  np.trace(m)
print("Sum of diagonal elements in matrix :")
print(result)
```
